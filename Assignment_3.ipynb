```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("netflix_titles.csv")

```

```python
df
```

```python
import matplotlib.pyplot as plt
# Plot histogram for 'Age' column
plt.hist(df['release_year'].dropna(), bins=20, edgecolor='black', color='skyblue')
plt.title('release_year')
plt.xlabel('release_year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for 'Pclass' (Passenger class)
plt.figure(figsize=(6, 4))
df['listed_in'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('listed_in Distribution')
plt.xlabel('listed_in')

plt.xlabel('country')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Box plot for 'Age' to visualize outliers and distribution
plt.figure(figsize=(6, 4))
plt.boxplot(df['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python

import seaborn as sns
# Pairplot to visualize relationships between numerical features
sns.pairplot(df[['release_year',"duration"]].dropna())
plt.show()
```

```python
# Count plot for 'Survived' (how many survived vs how many did not)
sns.countplot(x='release_year', data=df, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Violin plot for 'Age' across 'Pclass'
sns.violinplot(x='release_year', y='duration', data=df, palette='muted')
plt.title('release_year Distribution by duration')
plt.show()
```

```python
# Extract numeric duration from the 'duration' column
df['duration'] = df['duration'].str.extract('(\d+)').astype(float)

# Correlation heatmap
correlation_matrix = df[["release_year","duration"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
